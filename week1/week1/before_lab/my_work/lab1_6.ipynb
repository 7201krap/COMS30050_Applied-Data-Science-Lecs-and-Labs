Manual Query

```
# Specify the arguments
args = {
    'section': 'technology', 
    'order-by': 'newest', 
    'api-key': 'test', 
    'page-size': '100', # The number of results to return. 
    'q' : 'privacy%20AND%20data'
}

# Construct the URL
base_url = 'http://content.guardianapis.com/search'
url = '{}?{}'.format(
    base_url, 
    '&'.join(["{}={}".format(kk, vv) for kk, vv in args.items()])
)

# Make the request and extract the source
req = requests.get(url) 
print(url)
src = req.text
```

```
# load response by using json 
# check the difference between previous 'response'
response = json.loads(src)['response']
print(response.keys())
```

```
# sorted reponse's keys
response = json.loads(src)['response']
print(sorted(response.keys()))
```

```
# verify the status code 
assert response['status'] == 'ok'
```

```
print(json.dumps(response['results'], indent=2, sort_keys=True))
```

```
print(json.dumps(response['results'][0], indent=2, sort_keys=True))
```

```
for result in response['results']:
  print(result['webUrl'])
  print(result['webTitle'])
  print()
```
